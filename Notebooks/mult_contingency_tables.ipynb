```python
#calculate row-wise entropies:

#entropy for row 1:
e000 = (bp_ch[0][0])/(bp_ch["All"][0])
e001 = (bp_ch[1][0])/(bp_ch["All"][0])
re_00 = (-1)*(((e000)*np.log(e000)) + (e001*np.log(e001)))

#entropy for row 2:
e010 = (bp_ch[0][1])/(bp_ch["All"][1])
e011 = (bp_ch[1][1])/(bp_ch["All"][1])
re_01 = (-1)*(((e010)*np.log(e010)) + (e011*np.log(e011)))

#entropy for row3:
e100 = (bp_ch[0][2])/(bp_ch["All"][2])
e101 = (bp_ch[1][2])/(bp_ch["All"][2])
re_10 = (-1)*(((e100)*np.log(e100)) + (e101*np.log(e101)))

#entropy for row 4:
e110 = (bp_ch[0][3])/(bp_ch["All"][3])
e111 = (bp_ch[1][3])/(bp_ch["All"][3])
re_11 = (-1)*(((e110)*np.log(e110)) + (e111*np.log(e111)))

#calculate column wise entropies:

#col 0 entropy:
ce01 = (bp_ch[0][0])/(bp_ch[0][4])
ce02 = (bp_ch[0][1])/(bp_ch[0][4])
ce03 = (bp_ch[0][2])/(bp_ch[0][4])
ce04 = (bp_ch[0][3])/(bp_ch[0][4])
ce0 = (-1)*((ce01*np.log(ce01))+(ce02*np.log(ce02))+(ce03*np.log(ce03))+(ce04*np.log(ce04)))
print(ce0)

#col 1 entropy
ce11 = (bp_ch[1][0])/(bp_ch[1][4])
ce12 = (bp_ch[1][1])/(bp_ch[1][4])
ce13 = (bp_ch[1][2])/(bp_ch[1][4])
ce14 = (bp_ch[1][3])/(bp_ch[1][4])
ce1 = (-1)*((ce11*np.log(ce11)) + (ce12*np.log(ce12)) + (ce13*np.log(ce13)) + (ce14*np.log(ce14)))
print(ce1)

#row sum entropies:
r1 = -1*(((bp_ch["All"][0])/(bp_ch["All"][4]))*np.log((bp_ch["All"][0])/(bp_ch["All"][4])))
r2 = -1*(((bp_ch["All"][1])/(bp_ch["All"][4]))*np.log((bp_ch["All"][1])/(bp_ch["All"][4])))
r3 = -1*(((bp_ch["All"][2])/(bp_ch["All"][4]))*np.log((bp_ch["All"][2])/(bp_ch["All"][4])))
r4 = -1*(((bp_ch["All"][3])/(bp_ch["All"][4]))*np.log((bp_ch["All"][3])/(bp_ch["All"][4])))

#col sum entropies:
c1 = (-1)*((bp_ch[0][4])/(bp_ch["All"][4])+np.log((bp_ch[0][4])/(bp_ch["All"][4])))
c2 = (-1)*(((bp_ch[1][4])/(bp_ch["All"][4]))+np.log(((bp_ch[1][4])/(bp_ch["All"][4]))))

mce1 = ((r1+r2+r3+r4)/(c1+c2)) 
print(mce1)
```
